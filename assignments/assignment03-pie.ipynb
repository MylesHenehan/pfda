## Approach
In order to carry out this assignment, I will take the following steps:
1. Import the relevant libraries
2. Read in the data from CSV as a Pandas data frame
3. Extract Emails column from the data frame
4. Split up the emails to just get the domain
5. Use the series of domains to create a pie plot.

```python
# Importing useful libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Reading in CSV from URL
url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download'
df = pd.read_csv(url)

#just to check the name of the column I need
df.info()

emails = df['Email']

```

```python
# Extracting the domain from each email
splitemails = emails.str.split('@') #this splits each email into a list of 2 values, the part before the @ and after the @.
domains = splitemails.str[1]
print(domains)

count = domains.value_counts()
print(count)
```

```python
# Creating a pie chart
fig, ax = plt.subplots()
ax.pie(count, labels=count.index)
ax.set_title('Distribution of email domains')
plt.show()
```
