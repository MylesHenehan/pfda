This programme simulates 1000 individual battle rounds in Risk (3 attacker vs 2 defender) and plots the result.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
attackinglosses = []
defensivelosses = []

i = 0
while i < 1000:
    attack = np.sort(np.random.randint(low=1,high=7,size=3))
    defense = np.sort(np.random.randint(low=1, high=7, size=2))

    topattacker = attack[-1]
    topdefender = defense[-1]

    if topattacker > topdefender:
        defensivelosses.append(1)
    else:
        attackinglosses.append(1)


    secondattacker = attack[-2]
    seconddefender = defense[-2]


    if secondattacker > seconddefender:
        defensivelosses.append(1)
    else:
        attackinglosses.append(1)

    i = i + 1

defensivelosses = np.sum(np.array(defensivelosses))
attackinglosses = np.sum(np.array(attackinglosses))
    
print(f'Defensive Losses: {defensivelosses}')
print(f'Attacking Losses: {attackinglosses}')

```

```python
labels = [(f'Attacking losses ({attackinglosses})'), (f'Defensive Losses ({defensivelosses})')]
totals = [attackinglosses, defensivelosses]
colours = ['blue', 'red']

plt.bar(labels, totals, color=colours)
plt.title("Total Losses Comparison")
plt.ylabel("Total Losses")
plt.show()
```
