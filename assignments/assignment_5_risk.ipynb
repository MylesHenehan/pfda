This programme simulates 1000 individual battle rounds in Risk (3 attacker vs 2 defender) and plots the result.

## Approach
In order to carry out this assignment, after importing the relevant libraries, I will take the following steps:

For the simple version:
1. Use a while loop to simulate 1000 rounds and initiate lists to store the losses for each player.
2. For each round, generate random attack and defense rolls using the NumPy randint() function, and sort the values of the rolls in arrays in order of size.
3. Compare the attacker and defender rolls and add to the relevant losses list when each one loses.
4. Count the number of defensive and attacking losses, print the results and plot them.

For the more complicated version:
1. Define the sizes of the armies of each player.
2. Initialize the number of losses as 0
3. Use a while loop similar to the one above to simulate a series of rounds, but instead of playing out 1000 rounds, the condition of the loop will be that it continues until the losses outweigh the size of either of the armies.
4. To allow the armies to swap turns, I'll set it so that when the iterator is odd, it's the red army's turn, and when the iterator is even, it's the blue army's turn.
5. Once the first army loses all its troops and the loop breaks, I'll show the final results.





```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
attackinglosses = []
defensivelosses = []

i = 0
while i < 1000:
    attack = np.sort(np.random.randint(low=1,high=7,size=3))
    defense = np.sort(np.random.randint(low=1, high=7, size=2))

    topattacker = attack[-1]
    topdefender = defense[-1]

    if topattacker > topdefender:
        defensivelosses.append(1)
    else:
        attackinglosses.append(1)


    secondattacker = attack[-2]
    seconddefender = defense[-2]


    if secondattacker > seconddefender:
        defensivelosses.append(1)
    else:
        attackinglosses.append(1)

    i = i + 1

defensivelosses = np.sum(np.array(defensivelosses))
attackinglosses = np.sum(np.array(attackinglosses))
    
print(f'Defensive Losses: {defensivelosses}')
print(f'Attacking Losses: {attackinglosses}')

```

```python
labels = [(f'Attacking losses ({attackinglosses})'), (f'Defensive Losses ({defensivelosses})')]
totals = [attackinglosses, defensivelosses]
colours = ['blue', 'red']

plt.bar(labels, totals, color=colours)
plt.title("Risk: Attacking losses vs Defensive Losses")
plt.ylabel("Total Losses")
plt.show()
```

```python
labels = [(f'Red Losses ({redlosses})'), (f'Blue Losses ({bluelosses})')]
totals = [redlosses, bluelosses]
colours = ['red', 'blue']

plt.bar(labels, totals, color=colours)
plt.title("Risk: Red Army vs Blue Army")
plt.ylabel("Total Troop Losses")
plt.show()


```
