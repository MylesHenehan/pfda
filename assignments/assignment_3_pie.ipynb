## Approach
In order to carry out this assignment, I will take the following steps:
1. Import the relevant libraries
2. Read in the data from CSV as a Pandas data frame
3. Extract Emails column from the data frame
4. Split up the emails to just get the domain
5. Use the series of domains to create a pie plot.

```python
# Importing useful libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Reading in CSV from URL, and checking the column I need
url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download'
df = pd.read_csv(url)

df.info()
emails = df['Email']

```

```python
# Extracting the domain from each email
splitemails = emails.str.split('@') # this splits it into a series of lists with 2 elements, the part before the @ and the part after the @.
domains = splitemails.str.get(1) # here, .str.get() extracts the nth element of each list in the series, in this case the index being 1 (to get the domains).
print(domains)

count = domains.value_counts() # gets the number of occurences of each domain in the series
print(count)
```

```python
# Creating a pie chart with this count 
fig, ax = plt.subplots()
labels = [f'{domain} ({value})' for domain, value in zip(count.index, count)] # Used the information on zip() function from W3Schools to work out how I could show the actual counts in brackets (https://www.w3schools.com/python/ref_func_zip.asp)
ax.pie(count, labels=labels, autopct='%1.1f%%', shadow = True, startangle=90) # I added a few bells and whistles such as shadowing and changing the angle of the slices using the information from W3Schools on matplotlib pie charts (https://www.w3schools.com/python/matplotlib_pie_charts.asp)
ax.set_title('Distribution of email domains', fontsize=16, color='Black')

plt.show()
```
